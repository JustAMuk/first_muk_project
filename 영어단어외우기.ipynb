```python
while True:
    print("w는 쓰기, r은 읽고 시험하기, x는 나가기")
    command = input("모드를 선택하세요")
    
    if command == "x":
        break
    elif command == "w":
        num_write = 0
        word_list = []
        meaning_list = []
        while True:
            num_write +=1
            print("%d개째"%num_write)
            wordw = input("단어: ")
            meaningw = input("뜻: ")
            if wordw == "rtrt" or meaningw == 'rtrt' :
                print("이전거 지우기")
                word_list.remove(word_list[num_write - 2])
                meaning_list.remove(meaning_list[num_write - 2])
                num_write -=2
                continue
            if wordw == "" and meaningw == "":
                print("더는 없나봐?")
                break
            word_list.append(wordw)
            meaning_list.append(meaningw)
            if len(word_list) == 100:
                print("100개 했어")
                break
        Wmode = pd.DataFrame({"word":word_list,
                               "meaning":meaning_list})
        numb = int(input("몇번에 저장할거야? (0~9사이)"))
        Wmode.to_csv(location+name[numb], encoding = "ANSI", index = False)
        print("저장완료")            

    elif command == "r":
        filename = input("파일 이름을 말해줘")
        Rmode = pd.read_csv(location+filename+".csv", encoding = "ANSI")        
        wrong = pd.DataFrame({"word":[],"meaning":[]})
        while True:
            letter2 = Rmode.sample(n=len(Rmode), replace = False)
            letter3 = letter2.reset_index(drop = True)
            print("총 "+str(len(letter3)) + "개"  )
            
            save_wrong = input("틀린거 저장할까? y는 저장")
            for i in range(len(letter3)):
                print(str(i+1)+"개째")
                print(letter3.loc[i][0])
                while True:                    
                    a=input("계속하려면 p엔터, 이전거 저장하려면 rtrt")
                    if a == "rtrt":
                        wrong = wrong.append(letter3.loc[i-1])
                        print("이전거 저장 완료")
                        continue
                    elif a == "p":
                        break
                    else:
                        continue
                print(letter3.loc[i][1])
                while True:
                    b = input("저장하려면 k+엔터, 계속하려면 p+엔터, 나갈라면 qwqw엔터")
                    if b == "qwqw":
                        break
                    elif b == "k":
                        wrong = wrong.append(letter3.loc[i])
                        print("저장 완료")
                        break
                    elif b =="p":
                        break
                    else:
                        continue
                if b == "qwqw":
                    break
            wrong_beta = wrong
            print("y로 나가면 wrong이 0이 안되니까 참조")
            once = input("한번더 하고싶으면 y엔터")
            if once == "y":
                continue
            else:
                break
        if save_wrong == "y":
            wrong_alpha = pd.read_csv(location+wrong_filename, encoding = "ANSI")
            wrong = wrong_alpha.append(wrong)
            wrong.to_csv(location+wrong_filename, encoding = "ANSI", index = False)
            print("저장완료")
```
